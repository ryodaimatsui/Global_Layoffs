```python
# import dependencies
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import pandas as pd
```

```python
# Set up configuration to scrape dynamic website
url = 'https://companiesmarketcap.com/tech/largest-tech-companies-by-number-of-employees/'

chrome_driver_path = '/Users/peter/Downloads/chromedriver-mac-arm64/chromedriver'

service = Service(chrome_driver_path)

driver = webdriver.Chrome(service=service)

driver.get(url)
```

```python
# Retrieve respective xpaths for the columns needed.

companies = driver.find_elements(By.XPATH, '//tbody/tr/td[3]/div[2]/a/div[1]')
employees = driver.find_elements(By.XPATH, '//tbody/tr/td[4]')
price = driver.find_elements(By.XPATH, '//*[@id="cmkt"]/div[5]/table/tbody/tr/td[5]/div[1]')
percentage = driver.find_elements(By.XPATH, '//tbody/tr/td/div[2]/span')
countries = driver.find_elements(By.XPATH, '//*[@id="cmkt"]/div[5]/table/tbody/tr/td[8]/span')

companies_list = []

# Run a For Loop and place each row of data into a dictionary. Append that dictionary to the empty list.
for i in range(len(companies)):
    companies_dict = {'Company': companies[i].text,
                     'Employees': employees[i].text,
                     'Price': price[i].text,
                     'Percentage Change': percentage[i].text,
                     'Country': countries[i].text}
    companies_list.append(companies_dict)
```

```python
# Create DataFrame and view. 

companies_df = pd.DataFrame(companies_list)
companies_df
```

```python
# Countries text did not get extracted properly. Redo but this time use 'get_attribute' as the text may be hidden. 

countries = driver.find_elements(By.XPATH, '//*[@id="cmkt"]/div[5]/table/tbody/tr/td[8]/span')
countries_text = []

for country_element in countries:
    country_text = country_element.get_attribute('textContent')
    countries_text.append(country_text)
```

```python
# Check that the countries text were retrieved properly.
countries_text
```

```python
# Quit the driver.
driver.quit()
```

```python
# Update the DataFrame. 

companies_df['Country'] = countries_text

companies_df
```

```python
# Save the DataFrame as a CSV file to be used later on. 

companies_df.to_csv('top_tech_companies.csv', index=False)
```
